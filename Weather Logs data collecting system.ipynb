```python
import requests
from bs4 import BeautifulSoup
from datetime import datetime
import time
import csv
import pandas as pd
```

```python
with open('Measurements.csv','w') as file: #I made this file for storing data.
    fname=['Temp(C)','Wind(km/h)','Humidity(%)','Visibility(m)','Pressure(mb)','Time_Now(D/M/Y H:M:S)']
    writer=csv.DictWriter(file,fieldnames=fname)
    writer.writeheader()
    for i in range(10): #This for loop for refreshing the data 10 times from the web site.
        url="https://eg.freemeteo.com/weather/cairo/current-weather/location/?gid=360630&language=english&country=egypt"
        response=requests.get(url)
        soup=BeautifulSoup(response.text,'html.parser')
        temp=float(soup.find('div',attrs={'class':'temp'}).get_text().split('Â°')[0]) #I will get temperature degree in float.
        wind=float(soup.find('span',attrs={'class':'white'}).find('b').get_text().strip()) #I will get the wind spead in float.
        stats=soup.find('div',attrs={'class':'stats'}).get_text().split(' ') #I will get all stats in class stats.
        for el in stats: #I will take all data in the variable stats one by one by using for loop.
            if el.endswith('%'): #If el ends with % i will put it into humidity variable.
                humidity=el
            elif el.endswith('m'): #If el ends with m i will put it into visibility variable.
                visibility=el
            elif el.endswith('mb'): #If el ends with mb i will put it into pressure variable.
                pressure=el
        humidity=float(humidity.split('%')[0]) #I will take first index as float.
        visibility=float(visibility.split('m')[0]) #I will take first index as float.
        pressure=float(pressure.split('mb')[0]) #I will take first index as float.
        now=datetime.now() #Real time now, i will put it in now variable.
        time_now=now.strftime("%d/%m/%Y %H:%M:%S") #I used this format for making time arranged and i put it into time_now variable.
        output={'Temp(C)':temp, 'Wind(km/h)':wind, 'Humidity(%)':humidity, 'Visibility(m)':visibility, 'Pressure(mb)':pressure, 'Time_Now(D/M/Y H:M:S)':time_now}
        writer.writerow(output)
        print(f"Scraping done for {i+1} times at time:{time_now}")
        time.sleep(1) #I will make the data refreshed from the web site each second
```
